```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("http://www.bseindia.com/markets/equity/EQReports/mktwatchR.aspx?expandable=2&filter=loser*all$all$")
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python

table = soup.find( "table", {"id":"ctl00_ContentPlaceHolder1_grd1"} )
table
```

```python

tr = table.find_all('a', class_='tablebluelink')
company = []
for i in range(len(tr)):
    company.append(tr[i].text)
company
```

```python
tr = table.find_all('td', class_='TTRow_right')
LTP = []

for i in range(len(tr)):
    LTP.append(tr[i].text)
LTP
```

```python
Chg = []
for i in range(len(LTP)):
    if i+1 < len(LTP):
        Chg.append(LTP[i+1])
Chg
```

```python

PC = table.find_all('td', attrs={'style':'background-color:#DFDEDD;height:15px;width:40px;'})
PC
```

```python


PerChg = []
for i in range(len(PC)):
    PerChg.append(PC[i].text)
PerChg
```

```python
company.insert(0,'Company')
LTP.insert(0,'LTP')
Chg.insert(0,'Change')
PerChg.insert(0,'%Change')
rows = zip(company,LTP,Chg)
```

```python
import csv


with open('newfilePath2.csv', "w") as f:
    writer = csv.DictReader(f)
    for row in rows:
        writer.writerow(row)
```
